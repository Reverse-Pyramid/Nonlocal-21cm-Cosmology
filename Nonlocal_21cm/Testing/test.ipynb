### D results

```julia
PF.plot_d(10)
PFL.plot_d_LCDM!(10)
```

```julia
k = 100
z_nl, d_nl = PF.d_solve(k)
z_LCDM, d_LCDM = PFL.d_solve_LCDM(k)
d_ratio = (d_nl ./ d_LCDM) .^ 2 .- 1
plot(z_nl, d_ratio, title = "(D_nl / D_LCDM)^2 - 1", label = "k = $k", xlabel = "z", ylabel = "(D_nl / D_LCDM)^2 - 1")
```
